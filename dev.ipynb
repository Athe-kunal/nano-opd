```python
import io

def solve(input_str):
    input = io.StringIO(input_str)
    output = []
    t = int(input.readline())
    for _ in range(t):
        s = input.readline().strip()
        diff = sum(s[i] != "abc"[i] for i in range(3))
        output.append("YES" if diff <= 2 else "NO")
    return "\n".join(output) + "\n"

inputs = ['1\nabc\n', '3\nabc\nabc\nabc\n', '5\ncab\nacb\ncba\nbac\nbca\n', '6\nabc\nabc\nabc\nabc\nabc\nabc\n']
outputs = ['YES\n', 'YES\nYES\nYES\n', 'NO\nYES\nYES\nYES\nNO\n', 'YES\nYES\nYES\nYES\nYES\nYES\n']

for i, (inp, expected) in enumerate(zip(inputs, outputs)):
    result = solve(inp)
    status = "✅" if result == expected else "❌"
    print(f"Test {i}: {status}")
```
